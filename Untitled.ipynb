```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"person.tsv",sep='\t')
print("shape of person data is {}.".format(df.shape))
```

```python
#df.info()
df=df.rename(columns=str.lower)
df.head(5)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
count=df.isnull().sum()/df.shape[0]
percentage=100*((df.isnull().sum())/(df.shape[0]))
```

```python
table = pd.concat([count, percentage], axis=1).rename(columns = {0:"missing_count", 1:"missing_percentage"})
table
```

```python
table[table['missing_percentage'] > 40].round(2).transpose()
```

```python
to_drop = list(table[table['missing_percentage'] > 40].transpose().columns)  ## getting column names to drop
final_df= df.drop(columns = to_drop) 
```

```python
final_df
```

```python
final_df[final_df.duplicated()]# no duplicate rows 
```

```python
final_df["date_time"] = final_df["day"].apply(str) + "-" + final_df["month"].apply(str) + "-"+ final_df["year"].apply(str)
final=final_df.drop(columns=["day","month","year"])
final
```

```python
### bar plot function
def bar_plot(data, xlabel_, ylabel_, title_, num, size, color):
    """
    This function produces bar plot using pandas plot attribute
    """
    plot = data.value_counts()[0:num].plot(kind="bar", figsize=size, color=color)
    plt.xlabel(xlabel_, fontsize=15)
    plt.ylabel(ylabel_, fontsize=15)
    plt.title(title_, fontsize=15)
    plt.show()
```

```python
bar_plot(final["per_no"], "people", "count", "Total number of people  involved in each accident", 30, (10,5), color="blue")
```

```python
bar_plot(final["work_inj"], "people", "count", "Total number of people  involved in each accident", 30, (10,5), color="blue")
```

```python
bar_plot(final["veh_no"], "people", "count", "Total number of people  involved in each accident", 30, (10,5), color="blue")
```

```python
def bar_plot_unsorted(data, xlabel_, ylabel_, title_, num, size, color):
    """
    This function produces bar plot using pandas plot attribute
    """
    plot = data.value_counts().sort_index()[0:num].plot(kind="bar", figsize=size, color=color)
    plt.xlabel(xlabel_, fontsize=15)
    plt.ylabel(ylabel_, fontsize=15)
    plt.title(title_, fontsize=15)
    plt.show()
```

```python
bar_plot_unsorted(final["date_time"], "people", "count", "Total number of people  involved in each accident", 30, (10,5), color="blue")
```

```python
bar_plot_unsorted(final["hour"], "hours", "count", "accident count per hours", 24, (15,4), "black")
```
